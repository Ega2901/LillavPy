2. Числа изомерных алкильных углеводородных радикалов с п атомами углерода An
следующие можно вычислить сто, в уясь рекуррентной формулаланы в явном виде, а


```python
A = [1,1,1,2]

def calc_A(k):
    if k < 4:
        return A[k]
    
    for n in range(3,k):
        sum1 = sum(A[i] * A[j] * A[n-i-j] for i in range(n+1) for j in range(n-i+1))
        sum2 = sum(A[i] * A[n-2*i] for i in range((n//2)+1))
        sum3 = A[n//3]
        
        A.append((sum1 + sum2*3 + sum3*2) // 6)
    
    return A[k]

print(calc_A(int(input("Введите k:\n"))))
```

```python
from math import sqrt, pi
import random

def MonteCarlo(n):
    inside_circle = 0
    
    for _ in range(n):
        x = random.random()
        y = random.random()
        
        if  y <= sqrt(1 - x**2):
            inside_circle +=1
    
    pi_custom = (inside_circle / n) * 4
    return pi_custom

print("True_PI:", pi)
print("Ours_PI:", MonteCarlo(int(input("Enter n:\n"))))
```

```python
def get_element_symbols(formula):
    elements = []
    i = 0
    while i < len(formula):
        if formula[i].isupper():
            if i+1 < len(formula) and formula[i+1].islower():
                elements.append(formula[i:i+2])
                i += 2
            else:
                elements.append(formula[i])
                i +=1
        else:
            i += 1
    return elements

formula = input("Введите хим формулу:\n")
el_list = get_element_symbols(formula)
unique_el = set(el_list)
print("Список элементов:\n", el_list)
print("Уникальные элементы:\n", unique_el)

if unique_el == {"C", "H"}:
    print("Это УВ")
else:
    print("Это не УВ")
```
