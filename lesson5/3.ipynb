```python
with open('share/redwood-data.txt', 'r', encoding='utf-8') as f:
  data = [
      line.strip() for line in f if line.strip() and not line.startswith('#')
  ]

tallest_tree = None
widest_tree = None
max_height = 0
max_diametr = 0

for line in data:
  fields = line.split('\t')
  name = fields[0]
  height = float(fields[3])
  diametr = float(fields[2])
  
  if height > max_height:
    max_height = height
    tallest_tree = (name, height)
  if diametr > max_diametr:
    max_diametr = diametr
    widest_tree = (name, diametr)

print(f"Самое высокое дерево: {tallest_tree[0]} (высота {tallest_tree[1]} м)")
print(f"Cамое широкое дерево: {widest_tree[0]} (диаметр {widest_tree[1]} м)")
```
